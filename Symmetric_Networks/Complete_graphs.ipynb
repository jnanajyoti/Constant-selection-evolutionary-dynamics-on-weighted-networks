```python
# Drawing a complete graph with 8 nodes, dark blue colored nodes, and one thick edge

# Creating a complete graph with 8 nodes
G = nx.complete_graph(7)

# Setting edge weights: one edge thick (e.g., edge (0, 1)) and others normal
thick_edge = (3, 4)
for u, v in G.edges():
    if (u, v) == thick_edge or (v, u) == thick_edge:
        G[u][v]['weight'] = 8  # Thick edge
    else:
        G[u][v]['weight'] = 1  # Normal edges

# Generating positions for the graph
pos = nx.circular_layout(G)

# Drawing the graph
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=False, node_color='darkblue', node_size=700)

# Drawing edges with different thicknesses based on weights
for (u, v, d) in G.edges(data=True):
    nx.draw_networkx_edges(G, pos, edgelist=[(u, v)], width=d['weight'])

plt.savefig('complete_graph_hd.png', format='png', dpi=300)  # Save in HD
plt.show()

```
