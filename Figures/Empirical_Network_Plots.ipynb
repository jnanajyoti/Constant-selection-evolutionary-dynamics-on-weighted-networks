```python
def draw_star_graph_with_even_weighted_edges(G, ax, edge_weights):
    """
    Draws a star graph with weighted edges where the parallel lines are evenly spaced.
    The weighted edges are represented by multiple parallel lines, and all other edges
    are drawn normally. This version ensures a more professional and clean look.
    """
    # Position for star layout
    pos = nx.spring_layout(G)

    # Draw the nodes
    nx.draw_networkx_nodes(G, pos, ax=ax, node_color='darkblue')

    # Draw all edges, with special handling for weighted edges
    for (u, v, d) in G.edges(data=True):
        weight = d.get('weight', 1)  # Default weight is 1

        if weight in edge_weights:
            # Draw parallel edges for the specified weight
            for i in range(weight):
                # Evenly spaced offset for parallel edges
                offset = (i - (weight - 1) / 2.0) * 0.02
                new_pos = {node: (x + offset, y - offset) if node in [u, v] else (x, y) for node, (x, y) in pos.items()}
                nx.draw_networkx_edges(G, new_pos, edgelist=[(u, v)], ax=ax, edge_color='red', alpha=0.7, width=1.5)
        else:
            # Draw standard edges for weight 1
            nx.draw_networkx_edges(G, pos, edgelist=[(u, v)], ax=ax, edge_color='black')

    # Draw labels
    nx.draw_networkx_labels(G, pos, ax=ax, font_color='white')

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Draw the first graph (G3) with improved weighted edges
draw_star_graph_with_even_weighted_edges(G3, ax1, edge_weights=[2])
ax1.set_title("Star Graph (8 nodes, 2 edges of weight 2)")

# Draw the second graph (G4) with improved weighted edges
draw_star_graph_with_even_weighted_edges(G4, ax2, edge_weights=[3])
ax2.set_title("Star Graph (8 nodes, 4 edges of weight 3, 3 edges of weight 1)")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
def draw_centered_star_graph_with_even_weighted_edges(G, ax, edge_weights):
    """
    Draws a star graph with the central node in the center and other nodes arranged symmetrically around it.
    The weighted edges are represented by multiple parallel lines, and all other edges are drawn normally.
    """
    # Custom position for a centered star layout
    pos = {0: (0, 0)}  # Central node at (0, 0)
    angle = 2 * np.pi / (len(G.nodes()) - 1)
    for i in range(1, len(G.nodes())):
        pos[i] = (np.sin(i * angle), np.cos(i * angle))

    # Draw the nodes
    nx.draw_networkx_nodes(G, pos, ax=ax, node_color='darkblue')

    # Draw all edges, with special handling for weighted edges
    for (u, v, d) in G.edges(data=True):
        weight = d.get('weight', 1)  # Default weight is 1

        if weight in edge_weights:
            # Draw parallel edges for the specified weight
            for i in range(weight):
                # Evenly spaced offset for parallel edges
                offset = (i - (weight - 1) / 2.0) * 0.02
                new_pos = {node: (x + offset, y - offset) if node in [u, v] else (x, y) for node, (x, y) in pos.items()}
                nx.draw_networkx_edges(G, new_pos, edgelist=[(u, v)], ax=ax, edge_color='red', alpha=0.7, width=1.5)
        else:
            # Draw standard edges for weight 1
            nx.draw_networkx_edges(G, pos, edgelist=[(u, v)], ax=ax, edge_color='black')

    # Draw labels with white color
    nx.draw_networkx_labels(G, pos, ax=ax, font_color='white')

    # Remove the axis frame and set equal aspect
    ax.axis('off')
    ax.set_aspect('equal')

# Create a figure with two subplots, arranged vertically
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 12))

# Draw the first graph (G3) with the central node in the center
draw_centered_star_graph_with_even_weighted_edges(G3, ax1, edge_weights=[2])
ax1.set_title("$(N_1,N_2,w)=(5,2,2)$", y=1.05, fontsize=23)

# Draw the second graph (G4) with the central node in the center
draw_centered_star_graph_with_even_weighted_edges(G4, ax2, edge_weights=[3])
ax2.set_title("$(N_1,N_2,w)=(6,1,3)$", y=1.05, fontsize=23)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
